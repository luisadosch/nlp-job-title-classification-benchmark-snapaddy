```python
department_data = pd.read_csv("../../data/raw/department-v2.csv")
```

```python
department_data
```

```python
# show hoe many unique values are in the label column
department_data["label"].nunique()
# print all unique values in the label column
department_data["label"].unique()

# for each unique value in the label column, count how many times it appears
department_data["label"].value_counts()
```

```python
# for each unique value in the label column, show me one example text
for label in department_data["label"].unique():
    example = department_data[department_data["label"] == label]["text"].iloc[0]
    print(f"Label: {label}\nExample: {example}\n")
```
